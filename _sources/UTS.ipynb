

*   Import library pandas, numpy, nltk, dan re





```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
import re
import swifter
```



*   Load file dataset dari github menggungakan fungsi dari library pandas



```python
df = pd.read_csv("https://raw.githubusercontent.com/ThoriqFathu/prosaindata/main/dataset-baru.csv")
df
```



*   Memisahkan kolom target (kategori) dengan kolom abstrak yang nantinya akan dijadikan feature/attribut



```python
text = df['Abstrak']
text
```



*   Membuat fungsi untuk merubah seluruh huruf menjadi huruf kecil



```python
def case_folding(comment):
    comment = comment.lower()
    return comment
text = text.apply(case_folding)
text
```



---



*   Membuat fungsi untuk menghilangkan link/url, character, dan angka



```python
def linkNormalize(text):
    text = re.sub(r"\s—\s", "", text)
    text = re.sub(r"http\S+", "", text)
    text  = re.sub("@[A-Za-z0-9_]+","", text) #clenasing mention
    text = re.sub(r'http\S+', '', text) #cleansing url link
    text = re.sub("[^a-zA-Zï ]+"," ", text) #cleansing character
    return text
text = text.apply(linkNormalize)
text
```



*   Membuat fungsi untuk menghilangkan tanda baca



```python
symbols = ",!\"#$%&()*+-.…/:;<=>?@[\]^_`{|}~\n0987654321"
def cleaning_text(comment):
    for i in symbols:
        comment = np.char.replace(comment, i, ' ')
    return comment
text = text.apply(cleaning_text)
text
```



*   Membuat fungsi untuk menghilangkan kata yang tidak penting menggunakan library nltk



```python
nltk.download('stopwords')
list_stopwords = stopwords.words('indonesian')
def stopword_removal(comment):
    comment = comment.split()
    data = [word for word in comment if word not in list_stopwords]
    kalimat = ""
    for i in data:
      kalimat += i
      kalimat += " "
    return kalimat
```

```python
text = text.apply(stopword_removal)
```



*   Membuat fungsi untuk menjadikan kata ke kata dasar dengan library sastrawi



```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemmingNormalize(comment):
    comment = comment.split()
    data = []
    for term in comment:
        term = stemmer.stem(term)
        data.append(term)
    kalimat = ""
    for i in data:
      kalimat += i
      kalimat += " "
    return kalimat
text = text.swifter.apply(stemmingNormalize)
text
```



*   Melakukan ekstraksi fitur menggunakan fungsi TF-IDF dari library sklearn
*   Hasil ekstraksi fitur dijadikan dataframe menggunakan fungsi dari library pandas





```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_separate = tfidf_vectorizer.fit_transform(text)

df_tfidf = pd.DataFrame(
    tfidf_separate.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=df.index
)
X = df_tfidf.values
df_tfidf
```

*   Melakukan reduksi dimensi menggunakan metode PCA





```python
from sklearn.decomposition import PCA

pca = PCA(n_components =10)

X = pca.fit_transform(X)

pca.explained_variance_ratio_
```



*   Melakukan clustering dengan menggunakan metode K-means
*   Jumlah cluster sebanyak 2





```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```



*   Mengambil label hasil cluster
*   Merubah nilai 0 menjadi "RPL" dan 1 menjadi "KOMPUTASI"





```python
y_pred = kmeans.labels_


data = kmeans.labels_
res = []
for i in data:
    if i == 0:
        res.append("RPL")
    else:
        res.append("KOMPUTASI")
print(res)
```



*   Mengambil label sebenarnya dari dataset



```python
y_true = df['Kategori'].values.tolist()
print(df['Kategori'].values.tolist())
```



*   Melakukan analisa label hasil cluster dengan label sebenarnya



```python
n = len(y_true)
for i in range(n):
    print(res[i],"\t : ",y_true[i])
```



*   Menghitung akurasi hasil cluster



```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ",
    accuracy_score(y_true, res)*100)
```
