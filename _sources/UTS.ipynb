

*   Import library pandas



```python
import pandas as pd
```



*   Load file dataset dari github menggungakan fungsi dari library pandas



```python
df = pd.read_csv("https://raw.githubusercontent.com/ThoriqFathu/prosaindata/main/dataset-baru.csv")
df
```



*   Memisahkan kolom target (kategori) dengan kolom abstrak yang nantinya akan dijadikan feature/attribut



```python
text = df['Abstrak'].values.tolist()
Y = df['Kategori'].values
```



*   Melakukan ekstraksi fitur menggunakan fungsi TF-IDF dari library sklearn
*   Hasil ekstraksi fitur dijadikan dataframe menggunakan fungsi dari library pandas





```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_separate = tfidf_vectorizer.fit_transform(text)

df_tfidf = pd.DataFrame(
    tfidf_separate.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=df.index
)
X = df_tfidf.values
df_tfidf
```



*   Melakukan spilt dataset menjadi data training sebesar 80% dan data test sebesar 20% menggunkan library sklearn



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
    X, Y, test_size = 0.2, random_state = 100)
print("Jumlah Data training : ", len(X_train))
print("Jumlah Data test : ", len(X_test))
```



**Melakukan skenario uji coba dengan mereduksi fitur dengan n component sebanyak 2,4,6,8,10**

*   n component = 2





```python
from sklearn.decomposition import PCA

pca = PCA(n_components =2)

X_train_2 = pca.fit_transform(X_train)
X_test_2 = pca.transform(X_test)

pca.explained_variance_ratio_
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train_2, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test_2)
print (" KNN Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```
