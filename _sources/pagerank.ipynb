

*   imoport library networkx



```python
import networkx as nx
```



*   Membuat directed graph menggunakan fungsi pada library networkx



```python
G = nx.DiGraph([(1,2),(1,3),(1,4),(2,3),(3,4),(4,6),(5,6),(5,8),(6,7),(6,8),(7,8),(7,9),(9,1)])
```



*   Menampilkan graph yang telah di buat



```python
nx.draw(G, with_labels=True)
```

**Menerapkan Pagerank Simplified Algorithm**

*   Damping factor = 0.8
*   Max_iter = 100





```python
pr = nx.pagerank(G, alpha=0.8, max_iter=100)
pr
```



*   Meranking hasil perhitngan pagerank



```python
sorted(pr, key=pr.get, reverse=True)
```

**Kesimpulan**

*   Pada hasil perhitngan pagerank diatas dapat disimpulkan bahwasanya page 6 mendapatkan ranking tertinggi dibandingkan page yang lainnya
