**Seleksi Fitur**

*   Melakukan seleksi fitur dengan membedakan attribut dan kelas dimana X adalah fitur dan Y adalah kelas





```python
X = result.values[:, :4]
Y = result.values[:, 4]
```

**Membagi Data Training dan Data Test**

*   Membagi data training sebesar 80% dan data test sebesar 20% 





```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
    X, Y, test_size = 0.2, random_state = 100)
```

**Model Decision Tree**

*   Membuat model Decision Tree dengan menggunakan library sklearn





```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

**Akurasi**

*   Menghitung akurasi model dengan menggunakan fungsi accuracy_score pada library sklearn



```python
from sklearn.metrics import accuracy_score
print (" DecisionTree Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```

**Model K-Neares Neighbor**

*   Membuat model K-Neares Neighbor dengan menggunakan library sklearn



```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

**Akurasi**

*   Menghitung akurasi model dengan menggunakan fungsi accuracy_score pada library sklearn



```python
print (" KNN Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```
